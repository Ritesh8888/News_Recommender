```python
def opindia(uri):
    from bs4 import BeautifulSoup
    import requests
    url = uri
    data = requests.get(url)
    soup = BeautifulSoup(data.content)
    
    for i in soup.find_all("h1",class_={"tdb-title-text"}):
        title = i.text
    article = ""
    for i in soup.find_all("div",class_={"tdb-block-inner td-fix-index"}):
        article = article + str(i.text)
    date = article[0:23]
    date = date.replace(",","")
    return title,article,date
   
```

```python
opindia("https://www.opindia.com/2021/07/saket-gokhale-lakshmi-puri-deletes-tweets-delhi-high-court/")
```

```python
def findurls_inapage(baseurl):
    from bs4 import BeautifulSoup
    import requests
    url = baseurl
    data = requests.get(url)
    soup = BeautifulSoup(data.content)
    urlsinapage = []
    for i in soup.find_all("div",class_={"td_block_inner tdb-block-inner td-fix-index"}):
        for j in i.find_all("a",class_={"td-image-wrap"}):
            urlsinapage.append(j['href'])
    return urlsinapage
```

```python
findurls_inapage('https://www.opindia.com/2021/07/page/4/')
```

```python
baseurls = []

for i in range(2,10,1):
    url = "https://www.opindia.com/2021/07/page/"
    url = url + str(i) + "/"
    baseurls.append(url)
    
```

```python
baseurls
```

```python
allurls = []
for i in baseurls:
    allurls.extend(findurls_inapage(i))
```

```python
allurls
```

```python
titles=[]
articles=[]
dates=[]
for i in allurls:
    t,a,d = opindia(i)
    titles.append(t)
    articles.append(a)
    dates.append(d)

import pandas as pd
Q = pd.DataFrame([dates,titles,articles]).T
Q.columns=["dates","titles","articles"]
```

```python
import re
import nltk 

def sw(A3):
  from nltk.corpus import stopwords
  sw = stopwords.words("english")
  A4 = []
  for i in A3:
    if i not in sw:
      A4.append(i)
  return A4

#For Stemming
def stem(A4):
  from nltk.stem.porter import PorterStemmer
  ps = PorterStemmer()
  A5 = []
  for i in A4:
    A5.append(ps.stem(i))
  return A5


#For Lemmetization
def lem(A5):  
  from nltk.stem import WordNetLemmatizer
  wnlmt = WordNetLemmatizer()
  A6 = []
  for i in A5:
    A6.append(wnlmt.lemmatize(i))
  return A6
  

def tokenizer(A):
  from nltk.tokenize import word_tokenize
  A1 = word_tokenize(A)
  return A1

# text preprocessing
def rem_spl_chars(rev):
  rev = rev.lower()
  return re.sub("[^a-zA-Z0-9 ]","",rev)
  
  


```

```python
P1 = []
for i in Q.titles:
    for j in sw(stem(tokenize(rem_spl_chars(i)))):
        P1.append(j)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tv = TfidfVectorizer()
X = tv.fit_transform(P1).toarray()
```
